## Task 2: Missing Values (pre-built)

Check which columns have missing values and how many.

```python
# Count missing values per column
missing = df.isnull().sum() # Calculates the total number of missing (null) values for each column in the DataFrame.
missing_pct = (df.isnull().sum() / len(df) * 100).round(2) # Computes the percentage of missing values per column and rounds the result to two decimal places.

missing_report = pd.DataFrame({ # Creates a new DataFrame to organize the missing data statistics.
    'missing_count': missing, # Assigns the calculated counts of missing values to a column named 'missing_count'.
    'missing_pct': missing_pct # Assigns the calculated percentages of missing values to a column named 'missing_pct'.
})

# Show only columns with missing values
missing_report[missing_report['missing_count'] > 0] # Filters and displays only the rows in the report where the number of missing values is greater than zero.
```

```python
# Fill missing numerical values with median (robust to outliers)
numerical_cols = df.select_dtypes(include=[np.number]).columns # Identifies all columns in the DataFrame that contain numerical data types.
for col in numerical_cols: # Iterates through each identified numerical column.
    if df[col].isnull().sum() > 0: # Checks if the current numerical column contains any missing values.
        median_val = df[col].median() # Calculates the median value of the column to use as a replacement for nulls.
        df[col].fillna(median_val, inplace=True) # Replaces the missing values in the column with the calculated median directly in the original DataFrame.
        print(f"Filled {col} missing values with median: {median_val}") # Outputs a message confirming which column was filled and the median value used.

# Fill missing categorical values with mode
categorical_cols = df.select_dtypes(include=['object']).columns # Identifies all columns in the DataFrame that contain categorical (text-based) data types.
for col in categorical_cols: # Iterates through each identified categorical column.
    if df[col].isnull().sum() > 0: # Checks if the current categorical column contains any missing values.
        mode_val = df[col].mode()[0] # Retrieves the most frequently occurring value (mode) in the column.
        df[col].fillna(mode_val, inplace=True) # Replaces the missing values in the column with the calculated mode directly in the original DataFrame.
        print(f"Filled {col} missing values with mode: {mode_val}") # Outputs a message confirming which column was filled and the mode value used.

# Verify: no more missing values
print(f"\nTotal missing values remaining: {df.isnull().sum().sum()}") # Calculates and prints the sum of all remaining null values across the entire DataFrame to verify successful cleaning.
```

## Task 3: Remove Duplicates (pre-built)

```python
# Check for duplicates
n_duplicates = df.duplicated().sum() # Calculates the total number of duplicate rows existing in the DataFrame.
print(f"Duplicate rows found: {n_duplicates}") # Prints the count of identical rows identified in the dataset.

# Remove duplicates
if n_duplicates > 0: # Checks if the number of detected duplicates is greater than zero.
    df = df.drop_duplicates() # Deletes the duplicate rows from the DataFrame, keeping only the first occurrence of each.
    print(f"After removal: {df.shape[0]} rows remain") # Displays the new total row count after the cleaning process.
else: # Executes if no duplicate rows were found in the dataset.
    print("No duplicates to remove.") # Prints a message confirming that the dataset contains no redundant rows.
```
