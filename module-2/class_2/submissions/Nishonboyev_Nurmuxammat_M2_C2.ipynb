## Task 2: Missing Values (pre-built)

Check which columns have missing values and how many.

```python
# Count missing values per column
# This code checks for missing (null) values in the dataset.
# First, it counts how many missing values are in each column.
# Then, it calculates the percentage of missing values for each column.
# After that, it creates a new table (DataFrame) to organize this information.
# Finally, it shows only the columns that have missing values.

missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df) * 100).round(2)

missing_report = pd.DataFrame({
    'missing_count': missing,
    'missing_pct': missing_pct
})

# Show only columns with missing values
missing_report[missing_report['missing_count'] > 0]
```

```python
# Fill missing numerical values with median (robust to outliers)
# This code handles missing data in two steps: numerical and categorical.

# First, it finds all numerical columns (like int and float).
# Then, for each numerical column, it checks if there are missing values.
# If yes, it calculates the median (middle value) and fills missing values with it.
# Median is used because it is not affected by extreme values (outliers).
numerical_cols = df.select_dtypes(include=[np.number]).columns
for col in numerical_cols:
    if df[col].isnull().sum() > 0:
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)
        print(f"Filled {col} missing values with median: {median_val}")

# Fill missing categorical values with mode
# Next, it processes categorical columns (text data).
# For each column, it checks for missing values.
# If found, it calculates the mode (most common value) and fills missing values with it.
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        mode_val = df[col].mode()[0]
        df[col].fillna(mode_val, inplace=True)
        print(f"Filled {col} missing values with mode: {mode_val}")

# Verify: no more missing values
# Finally, it verifies that no missing values remain in the dataset.
# df.isnull().sum().sum() gives the total number of missing values.
print(f"\nTotal missing values remaining: {df.isnull().sum().sum()}")
```

## Task 3: Remove Duplicates (pre-built)

```python
# Check for duplicates
# This code checks if there are duplicate rows in the dataset.
# First, it counts how many duplicate rows exist using duplicated().
# Then, it prints the number of duplicates found.

n_duplicates = df.duplicated().sum()
print(f"Duplicate rows found: {n_duplicates}")

# Remove duplicates
# If duplicates exist, the code removes them using drop_duplicates().
# After removing, it prints the new number of rows in the dataset.
# If there are no duplicates, it prints a message saying everything is clean.
if n_duplicates > 0:
    df = df.drop_duplicates()
    print(f"After removal: {df.shape[0]} rows remain")
else:
    print("No duplicates to remove.")
```
