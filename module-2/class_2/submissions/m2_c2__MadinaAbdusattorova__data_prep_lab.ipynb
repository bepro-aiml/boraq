## Task 2: Missing Values (pre-built)

Check which columns have missing values and how many.

```python
# ---------------------------------------------------------
# TASK 2: Handling Missing Values
# ---------------------------------------------------------
# 1. Fill missing numerical values with the median (middle value).
# This is better than the mean because it is not affected by outliers.
# Count missing values in each column
missing = df.isnull().sum()
# Calculate percentage of missing values
missing_pct = (df.isnull().sum() / len(df) * 100).round(2)

# Combine into a report DataFrame
missing_report = pd.DataFrame({
    'missing_count': missing,
    'missing_pct': missing_pct
})

# Show only columns with missing values
missing_report[missing_report['missing_count'] > 0]
```

```python
# Select numerical columns (int, float)
numerical_cols = df.select_dtypes(include=[np.number]).columns
# Fill missing numerical values with median
# Median is robust to outliers
for col in numerical_cols:
    if df[col].isnull().sum() > 0:
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)
        print(f"Filled {col} missing values with median: {median_val}")

# Select categorical columns (object/string).
categorical_cols = df.select_dtypes(include=['object']).columns
# Fill missing categorical values with mode (most frequent value)
for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        mode_val = df[col].mode()[0]
        df[col].fillna(mode_val, inplace=True)
        print(f"Filled {col} missing values with mode: {mode_val}")

# Verify that no missing values remain
print(f"\nTotal missing values remaining: {df.isnull().sum().sum()}")
```

## Task 3: Remove Duplicates (pre-built)

```python
# Count duplicate rows
n_duplicates = df.duplicated().sum()
print(f"Duplicate rows found: {n_duplicates}")

# Remove duplicates if any exist
if n_duplicates > 0:
    df = df.drop_duplicates()
    print(f"After removal: {df.shape[0]} rows remain")
else:
    print("No duplicates to remove.")
```

## Task 5: Derived Features (TODO)

Create customer-level summary features. These are the building blocks for customer segmentation (Activity 4).

You need to create:
- **total_spending**: Total sales per customer
- **order_frequency**: Number of orders per customer
- **avg_order_value**: Average sales amount per order per customer

Hint: Use `df.groupby('Customer ID')` (or whatever the customer ID column is named).

```python
# Display all column names to identify correct fields
for col in df.columns:
    print(col)

# Total spending per customer (sum of Sales)
customer_spending = df.groupby('Customer ID')['Sales'].sum()
customer_spending.name = 'total_spending'

# Order frequency per customer (number of orders)
order_freq = df.groupby('Customer ID').size()
order_freq.name = 'order_frequency'

# Average order value per customer
avg_order = df.groupby('Customer ID')['Sales'].mean()
avg_order.name = 'avg_order_value'

# Combine all features into a single DataFrame
customer_summary = pd.concat(
    [customer_spending, order_freq, avg_order],
    axis=1
)

# Display first 10 rows
customer_summary.head(10)
```

## ð Task 5: Convert Data Types

## ð Task 6: Create Derived Features
> **Process:** Group data by Customer ID to calculate total spending, order frequency, and average order value, then export the final CSV files.

```python
# Calculate customer-level summary metrics
total_spending = df.groupby('Customer ID')['Sales'].sum().reset_index()
total_spending.columns = ['Customer ID', 'total_spending']

order_freq = df.groupby('Customer ID')['Order ID'].nunique().reset_index()
order_freq.columns = ['Customer ID', 'order_frequency']

avg_order = df.groupby('Customer ID')['Sales'].mean().reset_index()
avg_order.columns = ['Customer ID', 'avg_order_value']

# Merge the tables
customer_summary = total_spending.merge(order_freq, on='Customer ID').merge(avg_order, on='Customer ID')

# Save the cleaned data
df.to_csv('superstore_cleaned.csv', index=False)
customer_summary.to_csv('customer_summary.csv', index=False)

customer_summary.head(10)
```

In the final stage, a summary table was created for each customer, calculating total spending, order frequency, and average order value. The cleaned and processed data were then saved into new CSV files for further analysis
