```python
# Bu bo‘limda pandas va numpy kutubxonalari yuklanmoqda.
# pandas jadval bilan ishlash uchun, numpy esa hisob-kitoblar uchun ishlatiladi.
# Bu kutubxonalar keyingi barcha amallar uchun asos bo‘ladi.
import pandas as pd
import numpy as np
```

```python
# Bu yerda dataset CSV fayldan yuklanmoqda.
# encoding xatoliklarni oldini olish uchun berilgan.
# Natijada ma’lumotlar df ga saqlanadi.
url = "https://raw.githubusercontent.com/dsrscientist/dataset1/master/superstore.csv"
df = pd.read_csv(url, encoding='latin-1')
```

```python
# Datasetning birinchi 5 ta qatori ko‘rsatilmoqda.
# Bu orqali ustunlar va ma’lumotlar ko‘rinishi tekshiriladi.
# Tahlil boshlashdan oldin juda foydali.
df.head()
```

```python
# Bu yerda dataset o‘lchami olinmoqda.
# shape nechta qator va ustun borligini ko‘rsatadi.
# Dataset hajmini tushunishga yordam beradi.
df.shape
```

```python
# Dataset haqida umumiy texnik ma’lumot beriladi.
# Ustun turlari va bo‘sh qiymatlar ko‘rsatiladi.
# Tozalashdan oldin muhim bosqich.
df.info()
```

```python
# Statistik ma’lumotlar chiqarilmoqda.
# O‘rtacha, minimum, maksimum qiymatlar hisoblanadi.
# Sonli ustunlarni tahlil qilish uchun ishlatiladi.
df.describe()
```

```python
# Har bir ustundagi bo‘sh qiymatlar soni hisoblanmoqda.
# isnull() bo‘sh joylarni topadi.
# sum() ularni sanab beradi.
df.isnull().sum()
```

```python
# Bu yerda bo‘sh qiymatlar o‘chirilmoqda.
# dropna() null qatorlarni olib tashlaydi.
# Dataset toza holatga keladi.
df = df.dropna()
```

```python
# Datasetdagi barcha ustun nomlari chiqarilmoqda.
# Bu orqali qaysi ustunlar borligini ko‘rish mumkin.
# Keyingi ishlarda juda kerak bo‘ladi.
df.columns
```

```python
# Bu yerda faqat Sales ustuni tanlanmoqda.
# Ma’lum bir ustun bilan alohida ishlash mumkin.
# Bu tahlilni soddalashtiradi.
df["Sales"]
```

```python
# Bu bo‘limda filtr qo‘llanmoqda.
# Sales 100 dan katta bo‘lgan qatorlar olinadi.
# Faqat kerakli ma’lumotlar ajratiladi.
df[df["Sales"] > 100]
```

```python
# Yangi ustun yaratilmoqda.
# Profit va Sales asosida nisbat hisoblanmoqda.
# Bu qo‘shimcha tahlil uchun kerak.
df["Profit_ratio"] = df["Profit"] / df["Sales"]
```

```python
# Ma’lumotlar kategoriyalar bo‘yicha guruhlanmoqda.
# Har bir kategoriya uchun yig‘indi hisoblanadi.
# Bu umumiy natijalarni ko‘rsatadi.
df.groupby("Category").sum()
```

```python
# Ma’lumotlar saralanmoqda.
# Sales bo‘yicha kamayish tartibida joylashtiriladi.
# Eng katta qiymatlar yuqorida chiqadi.
df.sort_values(by="Sales", ascending=False)
```
