## 2. Dictionaries

A dictionary stores key-value pairs. Think of it as a single row in a database table -- keys are column names, values are the data.

```python
# --- EXAMPLE ---
# Dictionaries
# This script organizes structured data into a key-value format.
# It creates a comprehensive profile for a customer by mapping descriptive
# labels (keys) to specific pieces of information (values). This structure
# allows for quick and efficient data retrieval based on unique identifiers.

customer = {
    "id": "UZT-001234",
    "name": "Karimov Sherzod",
    "region": "Tashkent",
    "monthly_spend": 125000,
    "is_active": True,
    "plan": "Premium"
}

# Access values by key
# Accessing Dictionary Values
# This script demonstrates how to retrieve specific data by referencing its unique key.
# Unlike lists, which use numerical positions, dictionaries allow us to use
# descriptive labels to fetch information directly and intuitively.
print(f"Customer: {customer['name']}")
print(f"Region:   {customer['region']}")
print(f"Spend:    {customer['monthly_spend']} soum")

# Add a new key
# Modifying Dictionaries
# This script illustrates how to expand an existing dictionary by adding a new
# key-value pair. Dictionaries are flexible, allowing you to insert additional
# attributes—such as the customer's loyalty period—at any point during the program.
customer["tenure_months"] = 24
print(f"\nFull record: {customer}")
```

## 3. Functions

Functions take inputs, perform an operation, and return an output. They let you write reusable, modular code.

In ML, every step of a pipeline is typically a function: load_data(), clean_data(), train_model(), evaluate().

```python
# --- EXAMPLE: calculate_average ---
# Function Definition and Execution
# This script defines a reusable function designed to calculate the arithmetic
# mean of a numerical dataset. By encapsulating the logic within a function,
# we can perform this calculation multiple times on different lists without
# repeating the underlying code, making the program more efficient and modular.

def calculate_average(numbers):
    """Calculate the average of a list of numbers."""
    total = sum(numbers)
    count = len(numbers)
    return total / count

bills = [85000, 92000, 78000, 95000, 88000, 91000]
avg_bill = calculate_average(bills)
print(f"Average monthly bill: {avg_bill:,.0f} soum")
```

```python
# --- EXAMPLE: function with multiple returns ---
# Functions with Multiple Return Values
# This script defines a function that analyzes a dataset and provides several
# statistical insights at once. By returning a tuple of values, the function
# efficiently summarizes the data, allowing the caller to unpack the results
# into separate variables for further use.

def describe_bills(bills):
    """Return basic statistics for a list of bills."""
    avg = sum(bills) / len(bills)
    minimum = min(bills)
    maximum = max(bills)
    return avg, minimum, maximum

avg, low, high = describe_bills(bills)
print(f"Average: {avg:,.0f} | Min: {low:,.0f} | Max: {high:,.0f}")
```

## 4. Loops

Loops let you repeat an operation across a collection. In data work, you iterate over rows, columns, files, or parameter values constantly.

```python
# --- EXAMPLE: basic for loop ---
# Iterating with For Loops
# This script uses a 'for' loop to cycle through a collection of data.
# It automatically visits each item in the list one by one, assigning the
# current item to a temporary variable so that a specific action—such as
# printing a status message—can be performed for every element.

regions = ["Tashkent", "Samarkand", "Bukhara", "Fergana", "Namangan"]

for region in regions:
    print(f"Processing data for: {region}")
```

```python
# --- EXAMPLE: enumerate (index + value) ---
# enumerate gives you both the position and the value
# Iteration with Enumerate
# This script demonstrates how to track both the position and the value of items
# during a loop. By using the enumerate function, we can generate a counter
# automatically, which is more efficient and less error-prone than managing
# a manual counter variable outside the loop.

monthly_bills = [85000, 92000, 78000, 95000, 88000, 91000]

for i, bill in enumerate(monthly_bills):
    print(f"Month {i + 1}: {bill:,} soum")
```

```python
# --- EXAMPLE: loop with conditional ---
# Filtering Data with Conditional Loops
# This script processes a list of customer dictionaries and applies a logical
# filter. By nesting an 'if' statement inside a 'for' loop, we can examine
# each record and only display the data that meets a specific financial
# threshold—in this case, identifying high-value customers.

customers = [
    {"name": "Alisher", "spend": 200000},
    {"name": "Dilnoza", "spend": 45000},
    {"name": "Bobur",   "spend": 120000},
    {"name": "Malika",  "spend": 35000},
    {"name": "Rustam",  "spend": 180000},
]

print("High-value customers (spend > 100,000):")
for c in customers:
    if c["spend"] > 100000:
        print(f"  {c['name']}: {c['spend']:,} soum")
```
