## 1. Variables and Data Types

Python has four core data types you will use constantly:
- **int** -- whole numbers (subscriber count, age)
- **float** -- decimal numbers (bill amount, accuracy)
- **str** -- text (customer name, region)
- **bool** -- True or False (is_active, payment_on_time)
- **list** -- ordered collection (monthly bills, feature names)

```python
# --- EXAMPLES ---

# Integers
subscriber_count = 10_000_000 # Variable that containes the number of subscribers
age = 25 # Variable that stores the age

# subscriber_count is a variable that has 10_000_000 in it
# Floats
monthly_bill = 89_500.50 # Variable indicating the amount of monthly bill
accuracy = 0.95 # Variable for accuracy

# Strings
customer_name = "Abdullayev Jasur" # Variable storing the name of the customer
region = "Tashkent" # Variable for the name of the region

# Booleans
is_active = True # boolean representing if the customer is active of not
payment_on_time = False # boolean showing if the payment was completed on time or not

# Lists
monthly_bills = [85000, 92000, 78000, 95000, 88000] # List containing the amounts of monthly bills
regions = ["Tashkent", "Samarkand", "Bukhara", "Fergana"] # List that has different regions

# Check types
print(f"subscriber_count: {subscriber_count} (type: {type(subscriber_count).__name__})") # printing the output
print(f"monthly_bill:     {monthly_bill} (type: {type(monthly_bill).__name__})") # printing the output
print(f"customer_name:    {customer_name} (type: {type(customer_name).__name__})") # printing the output
print(f"is_active:        {is_active} (type: {type(is_active).__name__})") # printing the output
print(f"monthly_bills:    {monthly_bills} (type: {type(monthly_bills).__name__})") # printing the output
```

## 3. Functions

Functions take inputs, perform an operation, and return an output. They let you write reusable, modular code.

In ML, every step of a pipeline is typically a function: load_data(), clean_data(), train_model(), evaluate().

```python
# --- EXAMPLE: calculate_average ---

def calculate_average(numbers): # we are creating a function with numbers argument as a list
    """Calculate the average of a list of numbers."""
    total = sum(numbers) # variable the identifies the sum of the numbers in the inserted list
    count = len(numbers) # variable stores the length of the inserted list
    return total / count # returns the value obtained from dividing total(variable) by count(variable)

bills = [85000, 92000, 78000, 95000, 88000, 91000] # list
avg_bill = calculate_average(bills) # calling a function to calculate the average bill
print(f"Average monthly bill: {avg_bill:,.0f} soum") # printing the output
```

```python
# --- EXAMPLE: function with multiple returns ---

def describe_bills(bills): # a function with bills argument
    """Return basic statistics for a list of bills."""
    avg = sum(bills) / len(bills) # variable that contains the value received from sum divided by length of the bill(argument)
    minimum = min(bills) # variable with the minimum value in the inserted list
    maximum = max(bills) # variable with the minimum value in the inserted list
    return avg, minimum, maximum # returning the 3 values: average, minimum maximum

avg, low, high = describe_bills(bills) # destructuring the function.
print(f"Average: {avg:,.0f} | Min: {low:,.0f} | Max: {high:,.0f}") # printing the values
```

## 4. Loops

Loops let you repeat an operation across a collection. In data work, you iterate over rows, columns, files, or parameter values constantly.

```python
# --- EXAMPLE: basic for loop ---

regions = ["Tashkent", "Samarkand", "Bukhara", "Fergana", "Namangan"] # list that has regions

for region in regions: # loop that accesses every item in the list regions
    print(f"Processing data for: {region}") # printing every accessed item in the list regions
```

```python
# --- EXAMPLE: enumerate (index + value) ---
# enumerate gives you both the position and the value

monthly_bills = [85000, 92000, 78000, 95000, 88000, 91000]

for i, bill in enumerate(monthly_bills): # loop with the index - i and the accessed item in the list called bill
    print(f"Month {i + 1}: {bill:,} soum") # printing the index and then the item inside the list
```

```python
# --- EXAMPLE: loop with conditional ---

customers = [ # list with customers
    {"name": "Alisher", "spend": 200000}, # dictionary which has data of given people
    {"name": "Dilnoza", "spend": 45000},
    {"name": "Bobur",   "spend": 120000},
    {"name": "Malika",  "spend": 35000},
    {"name": "Rustam",  "spend": 180000},
]

print("High-value customers (spend > 100,000):") # print
for c in customers: # accessing every dictionary in the list called customers
    if c["spend"] > 100000: # creating the condition that states "if the spend value of accessed dictionary of the user exceeds 100000"
        print(f"  {c['name']}: {c['spend']:,} soum") # printing the name, spend values of the dictionary
```

```python
# TODO 2: Given the list of scores below, use enumerate to print each score
#         with its rank (starting from 1), and mark scores above 90 with " -- HIGH"
#
# Expected output:
#   Rank 1: 72
#   Rank 2: 95 -- HIGH
#   Rank 3: 88
#   ...

scores = [72, 95, 88, 61, 93, 77, 85, 91, 68, 99]

for rank, score in enumerate(scores, start=1): # loop that gives both the rank and the score value from the scores list
    if score > 90: # checking if the current score is greater than 90
        print(f"Rank {rank}: {score} -- HIGH") # printing the score with HIGH label for large values
    else: # this block works for scores that are not above 90
        print(f"Rank {rank}: {score}") # printing the rank and score without the HIGH label
```

```python
# TODO 3: Write a loop that computes the total and average of the scores list.
#         Do NOT use sum() or len() -- compute manually with a loop.
#         Print: "Total: <total>, Average: <average>"

total = 0 # variable that will store the sum of all scores
count = 0 # variable that will count how many scores are in the list

for score in scores: # loop that accesses every score in the scores list
    total += score # adding each score to the total variable
    count += 1 # increasing the number of processed scores by 1

average = total / count # calculating the average score using total and count
print(f"Total: {total}, Average: {average}") # printing the total and average values
```

```python
# --- TODO: Final exercise ---
#
# You have a dataset of 8 customers (list of dictionaries).
# Write code that:
# 1. Loops through each customer
# 2. Classifies them into tiers (use classify_customer or write inline logic)
# 3. Counts how many customers are in each tier
# 4. Prints a summary: "Premium: X, Standard: Y, Basic: Z"

customers_data = [
    {"id": 1, "name": "Anvar",   "monthly_spend": 210000},
    {"id": 2, "name": "Nilufar", "monthly_spend": 55000},
    {"id": 3, "name": "Sardor",  "monthly_spend": 180000},
    {"id": 4, "name": "Gulnora", "monthly_spend": 32000},
    {"id": 5, "name": "Timur",   "monthly_spend": 95000},
    {"id": 6, "name": "Madina",  "monthly_spend": 160000},
    {"id": 7, "name": "Otabek",  "monthly_spend": 47000},
    {"id": 8, "name": "Zarina",  "monthly_spend": 125000},
]

premium_count = 0 # variable that will count customers in the premium tier
standard_count = 0 # variable that will count customers in the standard tier
basic_count = 0 # variable that will count customers in the basic tier

for customer in customers_data: # loop that accesses every customer dictionary in customers_data
    tier = classify_customer(customer["monthly_spend"]) # classifying the customer by using monthly_spend
    if tier == "premium": # checking if the returned tier is premium
        premium_count += 1 # increasing the premium counter by 1
    elif tier == "standard": # checking if the returned tier is standard
        standard_count += 1 # increasing the standard counter by 1
    else: # this block works when the customer belongs to the basic tier
        basic_count += 1 # increasing the basic counter by 1

print(f"Premium: {premium_count}, Standard: {standard_count}, Basic: {basic_count}") # printing the final summary of all customer tiers
```
