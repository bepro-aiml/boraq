## 1. Variables and Data Types

- **int** -- whole numbers (subscriber count, age)
- **float** -- decimal numbers (bill amount, accuracy)

## 2. Dictionaries

A dictionary stores key-value pairs. Think of it as a single row in a database table -- keys are column names, values are the data.

```python
# --- EXAMPLE ---

# Defining a dictionary to store customer data as key-value pairs
customer = {
    "id": "UZT-001234",          # Unique identifier for the customer
    "name": "Karimov Sherzod",   # Customer's full name
    "region": "Tashkent",        # Location where the customer is based
    "monthly_spend": 125000,     # Amount spent per month (integer)
    "is_active": True,           # Current status (Boolean: active or not)
    "plan": "Premium"            # Subscription level/tier
}

# Access values by key
print(f"Customer: {customer['name']}")           # Retrieves and prints the value associated with 'name'
print(f"Region:   {customer['region']}")         # Retrieves and prints the value associated with 'region'
print(f"Spend:    {customer['monthly_spend']} soum") # Retrieves and prints the monthly spend with currency text

# Add a new key
customer["tenure_months"] = 24                   # Creates a new key-value pair for customer duration
print(f"\nFull record: {customer}")              # Prints the updated dictionary including the new entry
```

## 3. Functions

Functions take inputs, perform an operation, and return an output. They let you write reusable, modular code.

In ML, every step of a pipeline is typically a function: load_data(), clean_data(), train_model(), evaluate().

```python
# --- EXAMPLE: calculate_average ---

# Define a function named 'calculate_average' that accepts one parameter: 'numbers'
def calculate_average(numbers):
    """Calculate the average of a list of numbers.""" # Docstring: Explains what the function does
    total = sum(numbers)  # Calculates the sum of all elements in the list
    count = len(numbers)  # Gets the total number of items in the list
    return total / count  # Returns the result of dividing the total by the count

# Create a list of numeric values representing monthly bills
bills = [85000, 92000, 78000, 95000, 88000, 91000]

# Call the function with the 'bills' list and store the result in 'avg_bill'
avg_bill = calculate_average(bills)

# Print the formatted result; ':,.0f' adds a comma as a thousands separator and removes decimals
print(f"Average monthly bill: {avg_bill:,.0f} soum")
```

```python
# --- EXAMPLE: function with multiple returns ---

# Define a function that processes a list and returns three different metrics
def describe_bills(bills):
    """Return basic statistics for a list of bills.""" # Documentation string describing the function
    avg = sum(bills) / len(bills)       # Calculate the arithmetic mean of the bills
    minimum = min(bills)                # Identify the smallest value in the list
    maximum = max(bills)                # Identify the largest value in the list
    return avg, minimum, maximum        # Return all three values as a single tuple

# Call the function and "unpack" the returned tuple into three separate variables
avg, low, high = describe_bills(bills)

# Print the statistics using formatting to include thousands separators and zero decimal places
print(f"Average: {avg:,.0f} | Min: {low:,.0f} | Max: {high:,.0f}")
```

## 4. Loops

Loops let you repeat an operation across a collection. In data work, you iterate over rows, columns, files, or parameter values constantly.

```python
# --- EXAMPLE: basic for loop ---

# Define a list containing names of various geographical regions
regions = ["Tashkent", "Samarkand", "Bukhara", "Fergana", "Namangan"]

# Start a loop that iterates through each item in the 'regions' list
for region in regions:
    # In each iteration, the current item is assigned to the variable 'region' and printed
    print(f"Processing data for: {region}")
```

```python
# --- EXAMPLE: enumerate (index + value) ---
# enumerate gives you both the position and the value

# Create a list of monthly billing amounts (integers)
monthly_bills = [85000, 92000, 78000, 95000, 88000, 91000]

# Use enumerate to loop through the list, getting both the index (i) and the value (bill)
for i, bill in enumerate(monthly_bills):
    # Print the month number (index + 1, since Python starts at 0) and the formatted bill amount
    print(f"Month {i + 1}: {bill:,} soum")
```

```python
# --- EXAMPLE: loop with conditional ---

# Define a list of dictionaries, where each dictionary represents a customer and their spending
customers = [
    {"name": "Alisher", "spend": 200000},
    {"name": "Dilnoza", "spend": 45000},
    {"name": "Bobur",   "spend": 120000},
    {"name": "Malika",  "spend": 35000},
    {"name": "Rustam",  "spend": 180000},
]

# Print a header message to explain the filtered output
print("High-value customers (spend > 100,000):")

# Iterate through each customer dictionary (labeled as 'c') in the customers list
for c in customers:
    # Check if the value of the 'spend' key is greater than 100,000
    if c["spend"] > 100000:
        # If the condition is met, print the customer's name and their formatted spend amount
        print(f"  {c['name']}: {c['spend']:,} soum")
```
