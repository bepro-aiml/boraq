```python
# --- Task 2: Variables and Data Types ---
# Create variables
age = 22
gpa = 3.8
name = "Aziz"
is_student = True
scores = [85, 92, 78, 95, 88]

# Print variables with types
print(age, type(age))
print(gpa, type(gpa))
print(name, type(name))
print(is_student, type(is_student))
print(scores, type(scores))


```

```python
# --- Task 3: Function ---
# Function to calculate average
def calculate_average(numbers):
    total = sum(numbers)
    count = len(numbers)
    return total / count

# Test function
test1 = [80, 90, 70, 85, 95]
test2 = [55, 62, 48, 71]

print("Average 1:", calculate_average(test1))
print("Average 2:", calculate_average(test2))
```

```python
# --- Task 4: Loop with enumerate ---
cities = ["Tashkent", "Samarkand", "Bukhara", "Namangan", "Fergana"]

for index, city in enumerate(cities):
    print(f"Index {index}: {city}")
```

```python
# --- Task 5: Combine Everything ---
import random

# Create 10 random exam scores
exam_scores = [random.randint(0, 100) for _ in range(10)]
print("Scores:", exam_scores)

# Calculate average
avg = calculate_average(exam_scores)
print(f"Class average: {avg:.1f}")

# Check pass/fail
if avg >= 60:
    print("Result: PASS")
else:
    print("Result: FAIL")
```
