## 1. Variables and Data Types

Python has four core data types you will use constantly:
- **int** -- whole numbers (subscriber count, age)
- **float** -- decimal numbers (bill amount, accuracy)
- **str** -- text (customer name, region)
- **bool** -- True or False (is_active, payment_on_time)
- **list** -- ordered collection (monthly bills, feature names)

```python
# --- EXAMPLES ---

# Integers: Discrete whole numbers
subscriber_count = 10_000_000 # Total users
age = 25                      # User age

# Floats: Decimal numbers for precision
monthly_bill = 89_500.50      # Payment amount
accuracy = 0.95               # Probability rate

# Strings: Text-based data
customer_name = "Abdullayev Jasur" # Full name
region = "Tashkent"                # Location

# Booleans: Logic switches (True/False)
is_active = True        # Account status
payment_on_time = False # Payment status

# Lists: Collections of multiple items
monthly_bills = [85000, 92000, 78000, 95000, 88000] # History
regions = ["Tashkent", "Samarkand", "Bukhara", "Fergana"] # Service areas

# Debugging: Print values and their data types
print(f"subscriber_count: {subscriber_count} (type: {type(subscriber_count).__name__})")
print(f"monthly_bill:     {monthly_bill} (type: {type(monthly_bill).__name__})")
print(f"customer_name:    {customer_name} (type: {type(customer_name).__name__})")
print(f"is_active:        {is_active} (type: {type(is_active).__name__})")
print(f"monthly_bills:    {monthly_bills} (type: {type(monthly_bills).__name__})")
```

```python
# --- List operations ---

# Accessing: Get single items using their position (starts at 0)
print(f"First bill: {monthly_bills[0]}")  # Index 0 is the 1st item
print(f"Last bill:  {monthly_bills[-1]}") # -1 starts from the end

# Slicing: Extract a sub-range of the list
print(f"First 3 bills: {monthly_bills[:3]}") # Gets items from index 0 up to 3

# Length: Count total items in the list
print(f"Number of bills: {len(monthly_bills)}")

# Append: Add a new value to the very end of the list
monthly_bills.append(91000)
print(f"After append: {monthly_bills}")
```

## 2. Dictionaries

A dictionary stores key-value pairs. Think of it as a single row in a database table -- keys are column names, values are the data.

```python
# --- EXAMPLE ---

# Dictionary: Stores structured data using 'Key: Value' pairs
customer = {
    "id": "UZT-001234",          # Unique identifier
    "name": "Karimov Sherzod",   # String value
    "region": "Tashkent",        # String value
    "monthly_spend": 125000,     # Integer value
    "is_active": True,           # Boolean status
    "plan": "Premium"            # Category label
}

# Accessing: Retrieve data by using the specific Key name
print(f"Customer: {customer['name']}")
print(f"Region:   {customer['region']}")
print(f"Spend:    {customer['monthly_spend']} soum")

# Adding: Insert a new key and value into the existing dictionary
customer["tenure_months"] = 24
print(f"\nFull record: {customer}")
```

## 3. Functions

Functions take inputs, perform an operation, and return an output. They let you write reusable, modular code.

In ML, every step of a pipeline is typically a function: load_data(), clean_data(), train_model(), evaluate().

```python
# --- EXAMPLE: calculate_average ---

# Function: A reusable block that takes 'numbers' as input
def calculate_average(numbers):
    """Calculate the average of a list of numbers."""
    total = sum(numbers)   # Step 1: Add all numbers together
    count = len(numbers)   # Step 2: Count how many items exist
    return total / count   # Step 3: Divide and send back the result

# Data: A list of billing amounts
bills = [85000, 92000, 78000, 95000, 88000, 91000]

# Execution: Call the function and store the result in 'avg_bill'
avg_bill = calculate_average(bills)

# Output: Display formatted result (,:,.0f removes decimals and adds commas)
print(f"Average monthly bill: {avg_bill:,.0f} soum")
```

```python
# --- EXAMPLE: function with multiple returns ---

# Function: Calculates three statistics from a single list
def describe_bills(bills):
    """Return basic statistics for a list of bills."""
    avg = sum(bills) / len(bills)   # Metric 1: Arithmetic Mean
    minimum = min(bills)            # Metric 2: Lowest value in list
    maximum = max(bills)            # Metric 3: Highest value in list
    return avg, minimum, maximum    # Packaging results into a tuple

# Unpacking: Mapping the three returned values to three distinct variables
avg, low, high = describe_bills(bills)

# Output: Clean display using f-string formatting
print(f"Average: {avg:,.0f} | Min: {low:,.0f} | Max: {high:,.0f}")
```

## 4. Loops

Loops let you repeat an operation across a collection. In data work, you iterate over rows, columns, files, or parameter values constantly.

```python
# --- EXAMPLE: basic for loop ---

# List: A collection of string values representing locations
regions = ["Tashkent", "Samarkand", "Bukhara", "Fergana", "Namangan"]

# Loop: Iterates through each item in the 'regions' list
for region in regions:
    # Instruction: Executed once for every item in the list
    print(f"Processing data for: {region}")
```

```python
# --- EXAMPLE: enumerate (index + value) ---

# List: Historical billing data
monthly_bills = [85000, 92000, 78000, 95000, 88000, 91000]

# Loop: Uses enumerate to track both the counter 'i' and the data 'bill'
for i, bill in enumerate(monthly_bills):
    # Calculation: i + 1 is used because Python starts counting at 0
    print(f"Month {i + 1}: {bill:,} soum")
```

```python
# --- EXAMPLE: loop with conditional ---

# List of Dictionaries: Representing a customer database
customers = [
    {"name": "Alisher", "spend": 200000},
    {"name": "Dilnoza", "spend": 45000},
    {"name": "Bobur",   "spend": 120000},
    {"name": "Malika",  "spend": 35000},
    {"name": "Rustam",  "spend": 180000},
]

print("High-value customers (spend > 100,000):")

# Loop: Iterates through each customer dictionary 'c'
for c in customers:
    # Condition: Checks if the 'spend' value exceeds the 100k threshold
    if c["spend"] > 100000:
        # Execution: Only prints if the condition above is True
        print(f"  {c['name']}: {c['spend']:,} soum")
```
