## 2. Dictionaries

A dictionary stores key-value pairs. Think of it as a single row in a database table -- keys are column names, values are the data.

```python
# --- EXAMPLE ---

# This code shows how to store related data using a dictionary.
# A dictionary keeps information in key-value pairs.
# Each key (like "id", "name") has a value.
# This example stores information about one customer.
customer = {
    "id": "UZT-001234",
    "name": "Karimov Sherzod",
    "region": "Tashkent",
    "monthly_spend": 125000,
    "is_active": True,
    "plan": "Premium"
}

# Access values by key

# This code shows how to get values from a dictionary using keys.
# We use the key inside square brackets to access the needed value.
# Then, we print the customer's name, region, and monthly spend.
print(f"Customer: {customer['name']}")
print(f"Region:   {customer['region']}")
print(f"Spend:    {customer['monthly_spend']} soum")

# Add a new key

# This code shows how to add a new key-value pair to a dictionary.
# We add "tenure_months" with value 24 to the customer dictionary.
# This means the customer has been using the service for 24 months.
# Then, we print the full updated dictionary.
customer["tenure_months"] = 24
print(f"\nFull record: {customer}")
```

## 3. Functions

Functions take inputs, perform an operation, and return an output. They let you write reusable, modular code.

In ML, every step of a pipeline is typically a function: load_data(), clean_data(), train_model(), evaluate().

```python
# --- EXAMPLE: calculate_average ---

# This code defines a function to calculate the average of numbers,
# then uses that function to find the average monthly bill.

# The function takes a list of numbers as input.
# It calculates the total using sum(), counts items using len(),
# and returns the average (total divided by count).
def calculate_average(numbers):
    """Calculate the average of a list of numbers."""
    total = sum(numbers)
    count = len(numbers)
    return total / count
# Here we create a list of monthly bills.
bills = [85000, 92000, 78000, 95000, 88000, 91000]
# We call the function and store the result in avg_bill.
avg_bill = calculate_average(bills)
# Finally, we print the average value in a formatted way.
print(f"Average monthly bill: {avg_bill:,.0f} soum")
```

```python
# --- EXAMPLE: function with multiple returns ---

# This code defines a function that calculates basic statistics for a list of bills.
# It finds the average, minimum, and maximum values.
# Then, it returns all three results at the same time.
def describe_bills(bills):
    """Return basic statistics for a list of bills."""
    avg = sum(bills) / len(bills)
    minimum = min(bills)
    maximum = max(bills)
    return avg, minimum, maximum

# Here we call the function and store results in three variables.
avg, low, high = describe_bills(bills)
# Finally, we print the results in a formatted way.
print(f"Average: {avg:,.0f} | Min: {low:,.0f} | Max: {high:,.0f}")
```

## 4. Loops

Loops let you repeat an operation across a collection. In data work, you iterate over rows, columns, files, or parameter values constantly.

```python
# --- EXAMPLE: basic for loop ---

# This code shows how to loop through a list using a for loop.
# The loop goes through each region one by one.
# For every region, it prints a message.

regions = ["Tashkent", "Samarkand", "Bukhara", "Fergana", "Namangan"]

for region in regions:
    print(f"Processing data for: {region}")
```

```python
# --- EXAMPLE: enumerate (index + value) ---
# enumerate gives you both the position and the value
# This code shows how to use enumerate() to get both index and value from a list.
# enumerate() returns two things: the position (index) and the item (bill).
# We use i for the index and bill for the value.
# i + 1 is used because index starts from 0, but months start from 1.

monthly_bills = [85000, 92000, 78000, 95000, 88000, 91000]

for i, bill in enumerate(monthly_bills):
    print(f"Month {i + 1}: {bill:,} soum")
```

```python
# --- EXAMPLE: loop with conditional ---

# This code loops through a list of customers and checks a condition.
# For each customer, it looks at their spending amount.
# If the spend is greater than 100,000, the customer is considered high-value.
# Only those customers are printed.
customers = [
    {"name": "Alisher", "spend": 200000},
    {"name": "Dilnoza", "spend": 45000},
    {"name": "Bobur",   "spend": 120000},
    {"name": "Malika",  "spend": 35000},
    {"name": "Rustam",  "spend": 180000},
]

print("High-value customers (spend > 100,000):")
for c in customers:
    if c["spend"] > 100000:
        print(f"  {c['name']}: {c['spend']:,} soum")
```
