```python
# Har xil turdagi o'zgaruvchilarni yaratish
age = 22                 # int (butun son)
gpa = 3.85               # float (o'nlik son)
name = "Aziz"            # str (matn)
is_student = True        # bool (mantiqiy)
scores = [85, 92, 78]    # list (ro'yxat)

# Ularni turi bilan birga chiqarish
print(age, type(age))
print(gpa, type(gpa))
print(name, type(name))
print(is_student, type(is_student))
print(scores, type(scores))
```

```python
# Ro'yxatdagi sonlarning o'rtacha qiymatini hisoblaydigan funksiya
def calculate_average(numbers):
    total = sum(numbers)
    count = len(numbers)
    return total / count

# Funksiyani tekshirib ko'rish
test1 = [80, 90, 70, 85, 95]
test2 = [55, 62, 48, 71]

print("Average 1:", calculate_average(test1))
print("Average 2:", calculate_average(test2))
```

```python
# Shaxarlar ro'yxati
cities = ["Tashkent", "Samarkand", "Bukhara", "Namangan", "Fergana"]

# Har bir shahar va uning indeksini chiqarish
for index, city in enumerate(cities):
    print(f"Index {index}: {city}")
```

```python
import random

# 1. 10 ta tasodifiy imtihon ballarini yaratish (0 dan 100 gacha)
exam_scores = [random.randint(0, 100) for _ in range(10)]
print("Scores:", exam_scores)

# 2. Yuqorida yaratgan funksiyamiz orqali o'rtacha ballni hisoblash
avg = calculate_average(exam_scores)
print(f"Class average: {avg:.1f}")

# 3. O'tish chegarasini tekshirish (chegara: 60)
if avg >= 60:
    print("Result: PASS")
else:
    print("Result: FAIL")
```
