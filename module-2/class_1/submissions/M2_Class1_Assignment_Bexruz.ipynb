```python
#task3 ortacha qiymatni hisoblaydigan funksiya
def calculate_average(numbers):
  # sum() funksiyasi ro'yxatdagi hamma sonlarni qo'shib beradi
  # len() funksiyasi ro'yxatda nechta son borligini aniqlaydi
    average = sum(numbers) / len(numbers)
    return average

# funksiyani tekshiramiz
test_list = [10, 20, 30, 40, 50]
print("Natijasi:", calculate_average(test_list))
```

```python
test1 = [10,50,100,40,100]
test2 = [75,60,90,50]
print("Average 1:", calculate_average(test1))
print("Average 2:", calculate_average(test2))
```

```python
#task4 shaxarlar royxati va enumerate() operatori bilan ishlash
# enumerate() bizga ham indeksni (i), ham qiymatni (city) qaytaradi
cities_and_vegetables = ["Tashkent", "New York", "Seoul", "Washington", "Moscow","Apple","Banana","Pineapple"]
for i, city in enumerate(cities_and_vegetables):
      print(f"Index {i}: {city}")
```

```python
import random

# 1) 10 ta tasodifiy imtihon ballaridan iborat ro'yxat yaratamiz
exam_scores = [random.randint(0, 100) for _ in range(10)]
print("Imtihon ballari:", exam_scores)

# 2) Task 3 dagi funksiyamizni chaqirib, o'rtacha ballni hisoblaymiz
avg = calculate_average(exam_scores)
print(f"O'rtacha ball: {avg:.1f}") # :.1f nuqtadan keyin bitta raqam chiqaradi

# Shartli operator orqali PASS yoki FAIL ekanligini aniqlaymiz
if avg >= 60:
    print("Result: PASS")
else:
    print("Result: FAIL")
```
